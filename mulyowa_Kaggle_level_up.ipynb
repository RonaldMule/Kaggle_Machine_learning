```python
melbourne_data= melbourne_data.dropna(axis=0)
```

```python
y = melbourne_data.Price
```

```python
melbourne_feature = ['Rooms','Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
```

```python
x = melbourne_data[melbourne_feature]
```

```python
x.describe()
```

```python
x.head()
```

```python
from sklearn.tree import DecisionTreeRegressor
#here we are defining the decision tree and the fitting it to the features
#Define a model(melbourne_mode) and specify a number of random_state to ensure same results in each run
melbourne_model = DecisionTreeRegressor(random_state=1)

#fit the model to the features
melbourne_model.fit(x,y)
```

```python
#predicting the prices for houses
print("Making pridictions for the following 5 houses:")
print(x.head())
print ("the predictions are")
print(melbourne_model.predict(x.head()))
```
